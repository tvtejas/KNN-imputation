```python
iris = iris.dropna().drop('Unnamed: 0', axis = 1)
```

```python
y = iris.Species
x = iris[["Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.5, random_state = 123, stratify = y)
```

```python
y_predict=y_test.copy()
y_predict=y_predict.reset_index()
```

```python
def euclidean(a,b):
    distance = np.sqrt(np.sum((np.array(a)-np.array(b))**2))   
    return(distance)
```

## KNN

```python
def knn(x_train,x_test,y_train):
    minm=[]
    for i in range(0,x_test.shape[0]):
        dist=[]
        a=x_test.iloc[i:i+1]
        for j in range(0,x_train.shape[0]):
            b=x_train[j:j+1]
            dist.append(euclidean(a,b))
        minm.append(dist.index(min(dist)))
    y_pred=[]
    for i in minm:
        y_pred.append(y_train[y_train.index[i]])
    y_predict["k_1"]=y_pred
    return(y_predict)
```

```python
knn(x_train,x_test,y_train)
```

## KNN Classifier

```python
def knnclassifier(x_train,x_test,y_train, k):
    from scipy import stats
    minm=[]
    for i in range(0,x_test.shape[0]):
        dist=[]
        a=x_test.iloc[i:i+1]
        for j in range(0,x_train.shape[0]):
            b=x_train[j:j+1]
            dist.append([euclidean(a,b),j]) # calculating Euclidean distance and storing along with index
        dist.sort() # sorting distance
        minm.append(dist[:k]) # taking first k minimum distance 
    y_pred=[]
    for i in range(0,x_train.shape[0]):
        for j in range(0,k):
            pred="" # creating empty string to store predicted value k times
            pred=pred+y_train[y_train.index[minm[i][j][1]]]
        y_pred.append(stats.mode(pred)) # taking mode of k predicted value
    y_predict["Predicted"]=y_pred#creating new column and storing output value 
    
    return(y_predict)
```

```python
knnclassifier(x_train,x_test,y_train,int(input("Enter the value of k: ")))
```
